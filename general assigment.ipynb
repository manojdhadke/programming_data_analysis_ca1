```python
def get_fraction(numerator, denominator):  
    for divisor in range(2, 100):
        if (numerator % divisor == 0 and denominator % divisor == 0):
            return(get_fraction(numerator / divisor, denominator / divisor))
    return('{}/{}'.format(numerator, denominator))

try:
    numerator = int(input("Enter numerator decimal number: "))
    denominator = int(input("Enter denominator decimal number: "))

    # Displaying number in fraction 
    print (f"Smallest form without any common factor between {numerator} and {denominator} is", get_fraction(numerator, denominator))
except ValueError:
    print("\033[91mPlease enter valid decimal number. For Ex: 10, 22, 37, 43... so on")

```

```python
# Day names
days = ("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday") 

# Function to check whether year is leap year or not
def is_leap_year(year): 
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    else:
        return False

#Calculate the Julian calender day for a given year.
def calculate_julian_calender_day(year, month = 1, day = 1):
    if month <= 2:
        year -= 1
        month += 12
    if is_leap_year(year):
        feb_offset = 1
    else:
        feb_offset = 2
    return day + ((153*month - 457)//5) + (365*year) + (year//4) - 32083 + feb_offset

try:
    year = int(input("Enter the year(Y, YY, YYY, YYYY): ")) 
    
    # To handle Gregorian calendar dates from 1752 to .....
    if(year > 1752):
        # As per year and day: 1st Jan 2023 is a Sunday
        year_ref = 2023 
        day_index = 0
    
        years_diff = abs(year - year_ref)
        if year >= year_ref: 
            for index in range(year_ref, year_ref + years_diff): 
                if is_leap_year(index): 
                    day_index += 2 
                else: 
                    day_index += 1 
        else: 
            for index in range(year, year+years_diff): 
                if is_leap_year(index): 
                    day_index -= 2         
                else: 
                    day_index -= 1 
        day_index = day_index % 7 
        print(f"The day of date(Year-Month-Date) {year}-1-1 must be", days[day_index])

    # To handle Julian calendar dates from 1 to 1752
    elif(year > 9): 
        #Calculate the day of the week for a given date in the Julian calendar.
        day_index = calculate_julian_calender_day(year) % 7
        print(f"The day of date(Year-Month-Date) {year}-1-1 must be", days[day_index])

    # Handling descret values from 1st year to 9th year            
    elif(year == 9):     
        print(f"The day of date(Year-Month-Date) 9-1-1 must be", days[2])
    elif(year == 8):     
        print(f"The day of date(Year-Month-Date) 8-1-1 must be", days[0])
    elif(year == 7):     
        print(f"The day of date(Year-Month-Date) 7-1-1 must be", days[6])
    elif(year == 6):     
        print(f"The day of date(Year-Month-Date) 6-1-1 must be", days[5])
    elif(year == 5):     
        print(f"The day of date(Year-Month-Date) 5-1-1 must be", days[4])
    elif(year == 4):     
        print(f"The day of date(Year-Month-Date) 4-1-1 must be", days[2])
    elif(year == 3):     
        print(f"The day of date(Year-Month-Date) 3-1-1 must be", days[1])
    elif(year == 2):     
        print(f"The day of date(Year-Month-Date) 2-1-1 must be", days[0])
    elif(year == 1):     
        print(f"The day of date(Year-Month-Date) 1-1-1 must be", days[6])
    else:
        print("\033[91mPlease enter valid year. It should not be 0 or negative number. For Ex: 1, 99, 199, 1754, 1755, 2050...")
    
except ValueError:
    print("\033[91mPlease enter valid year. Make sure it should be greater than 0.  For Ex: 1, 99, 199, 1754, 1755, 2050...")
```

```python
def get_random_number(min_val, max_val, seed):
    # Used linear congruential generator (LCG) method for generating pseudo-random numbers
    return min_val + (1103515245 * seed + 12345) % (2 ** 31) % (max_val - min_val + 1)

def get_binary_search_guesses(L, H, t):
    guesses = 0 # Initiallizing guesses to zero
    while L <= H:
        guesses += 1
        g = (L + H) // 2
        # Possible 3 condition: g < t; g = t; g > t 
        if g == t:
            return guesses
        elif g < t:
            L = g + 1
        else: #g > t 
            H = g - 1
    return guesses
        
def get_random_binary_search_guesses(L, H, t, seed):
    guesses = 0        
    while L <= H:
        guesses += 1
        seed += 1
        # In comparison with Binary Search is g will be choosen randomly with function get_random_number() 
        g = get_random_number(L, H, seed)
        # Possible 3 condition: g < t; g = t; g > t 
        if g == t:
            return guesses
        elif g < t:
            L = g + 1    
        else: #g > t 
            H = g - 1
    return guesses

def binary_search(number):
    total_guesses = 0
    for t in range(1, number+1):
        total_guesses += get_binary_search_guesses(1, number, t)
    return total_guesses / number

def random_binary_search(number):
    total_guesses = 0
    seed = 96 # Default value
    try:
        seed = int(input("Enter the seed number to generate random number(1-100) or ignore this prompt:"))
        # It can be any number but for program efficiency, decided range of 1 - 100
        if(seed < 1 or seed > 101):
            print("\033[91mSeed number is not in specified range thus using default value 96\033[0m")
    except ValueError:
        print("\033[91mSince seed number is not correct hence using default value 96\033[0m")

    for t in range(1, number+1):
        total_guesses += get_random_binary_search_guesses(1, number, t, seed)
        seed+=1
    return total_guesses / number


# Following will require extra time to compute, thus kept commented simple code to debug and test
number = 10**10
# number = 2**3

binary_search_guesses = binary_search(number);
random_binary_search_guesses = random_binary_search(number);

print("\nBinary Search Guesses:esses: ", binary_search_guesses)
print("Random Binary Search Guesses:", random_binary_search_guesses)

search_difference = round((random_binary_search_guesses - binary_search_guesses), 8)
print("\nFound R(10^10) − B(10^10) rounded to 8 decimal places:\033[1m", search_difference, "\033[0m")
```

```python
counts = {"Even": 0, "Odd": 0, "Prime": 0, "Palindrome": 0}

# Checks given number is prime or not
def is_prime(number):
    if number < 2:
        return False
    for index in range(2, int(number**0.5)+1): #Calculating square root of number rounded to the next integer to save CPU time
        if number % index == 0:
            return False
    return True

# Checks given number is palidrome or not
def is_palindrome(number):
    # str(number)[::-1] will reverse the number for ex: 18 => 81
    return str(number) == str(number)[::-1]

for index in range(1, 1001):
    # Checks given number is divisible by 2 so that it can be Even or not
    if index % 2 == 0:
        counts["Even"] += 1 # Increasing the count of Even
    else:
        counts["Odd"] += 1 # Increasing the count of Odd
        
    if is_prime(index):
        counts["Prime"] += 1 # Increasing the count of Prime
        
    if is_palindrome(index):
        counts["Palindrome"] += 1 # Increasing the count of Palindrome

print("Total counts of Even, Odd, Prime and Palindrome numbers is:", counts, "\n")
print("Total counts of Even numbers is", counts["Even"])
print("Total counts of Odd numbers is", counts["Odd"])
print("Total counts of Prime numbers is", counts["Prime"])
print("Total counts of Palindrome numbers is", counts["Palindrome"])
```

```python
class Triangle:
    def __init__(self, side1, side2, side3):
        # Initializing variables
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def __is_valid(self):
        # As a general rule: Any side of a triangle must be shorter than the other two sides added together
        # To do so, checking if the sum of any two sides is greater than the third side
        return (self.__side1 + self.__side2 > self.__side3) and \
               (self.__side2 + self.__side3 > self.__side1) and \
               (self.__side1 + self.__side3 > self.__side2)

    def get_type(self):
        if not self.__is_valid():
            return "Invalid triangle"
        elif self.__side1 == self.__side2 == self.__side3:
            return "Equilateral triangle"
        elif self.__side1 == self.__side2 or self.__side1 == self.__side3 or self.__side2 == self.__side3:
            return "Isosceles triangle"
        else:
            return "Scalene triangle"

triangle_one = Triangle(3, 4, 5)
print("Side: 3, 4, 5 and Type of triangle is:\033[1m",triangle_one.get_type(), "\033[0m")  # Output: Scalene triangle

triangle_two = Triangle(3, 3, 3)
print("Side: 3, 3, 3 and Type of triangle is:\033[1m", triangle_two.get_type(), "\033[0m")  # Output: Equilateral triangle

triangle_three = Triangle(3, 3, 5)
print("Side: 3, 3, 5 and Type of triangle is:\033[1m", triangle_three.get_type(), "\033[0m")  # Output: Isosceles triangle

triangle_four = Triangle(1, 2, 3)
print("Side: 1, 2, 3 and Type of triangle is:\033[1m\033[91m", triangle_four.get_type(), "\033[0m")  # Output: Invalid triangle
```
